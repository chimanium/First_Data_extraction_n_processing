```python
# Exercise 2
# Converting GDP column to billion USD
df['GDP in Millions(USD)'] = df['GDP in Millions(USD)'].str.replace(',', '')
df['GDP in Millions(USD)']= df['GDP in Millions(USD)'].str.replace('\n', '')
df['GDP in Millions(USD)'] = df['GDP in Millions(USD)'].str.replace('—', '0')


```

```python
df
```

```python
df['GDP in Millions(USD)'] = df['GDP in Millions(USD)'].astype(int)
```

```python
df.iloc[3:11, 1]
```

```python
df3 = df.iloc[1:11, :]
```

```python
df3
```

```python
# Converting from Millions to Billions

df3[['GDP in Millions(USD)']] = df3[['GDP in Millions(USD)']]/1000

df3

```

```python
#rounding up
df3[['GDP in Millions(USD)']]= np.round(df3[['GDP in Millions(USD)']], 2)
```

```python
#rename columns

df3.rename(columns = {'GDP in Millions(USD)': 'GDP in Billions(USD)'}, inplace = True)
```
